```python
import os
from PIL import Image
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
from matplotlib.image import imread
import numpy as np
```

```python
def unpickle(file):
    import _pickle as cPickle
    dt = cPickle.load(file, encoding='latin1')
    return dt
```

```python
img = imread("./face.jpg")
imgplot = plt.imshow(img)
plt.show()
```

```python
img.shape
```

```python
int_img = np.zeros(img.shape)

for i, row in enumerate(img):
    for j, column in enumerate(img):
       
        if i is not 0 and j is not 0:
            int_img[j,i] = img[j,i] + img[j - 1,i]
        elif j is not 0 and j is 0:
            int_img[j,i] = img[j,i] + img[j - 1,i]
        elif i is not 0 and i is 0:
            int_img[j,i] = img[j,i] + img[j,i - 1]
```

```python
imgplot = plt.imshow(int_img)
plt.show()
imgplot = plt.imshow(img)
plt.show()
```
