```python
import cv2
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
from matplotlib.image import imread
import numpy as np
```

```python
snap = cv2.imread("./face.jpg")[::-1]
print(snap[0][:2])
for idx, row in enumerate(snap):
    for index, item in enumerate(row):
        arr = row[index]
        steps = 2
        snap[idx][index] = np.roll(row[index], steps)
print(snap[0][:2])
```

```python
imgplot = plt.imshow(snap)
plt.show()
```

```python
clf = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_frontalface_default.xml')
```

```python
gray = cv2.cvtColor(snap, cv2.COLOR_RGB2GRAY)
imgplot = plt.imshow(gray)
plt.show()
```

```python
faces = clf.detectMultiScale(gray, 1.3, 5)
```

```python
faces
```

```python
for (x,y,w,h) in faces:
        cv2.rectangle(snap,(x,y),(x+w,y+h),(255,0,0),2)
```

```python
rect = cv2.rectangle(snap, (x,y),(x+w,y+h),(255,0,0),2)
imgplot = plt.imshow(rect)
plt.show()
```
